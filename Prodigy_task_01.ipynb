```python
df= pd.read_csv(r"C:\Users\HP\Downloads\heart.csv")
df.head()
```

```python
df.isnull().sum()
df = df.dropna()
df = df.drop_duplicates()
```

```python
df.head()
```

```python
df.to_csv('cleaned_dataset.csv', index=False)
df = pd.read_csv('cleaned_dataset.csv')
df.head()
```

```python
#  IMPORT NECESSARY LIBRARIES 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot Age Distribution Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=10, kde=True, color='blue', alpha=0.7)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution in Heart Disease Dataset")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plot Gender Distribution Bar Chart
plt.figure(figsize=(8, 5))
sns.countplot(x=df['sex'].map({1: "Male", 0: "Female"}), palette="viridis")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Distribution in Heart Disease Dataset")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
